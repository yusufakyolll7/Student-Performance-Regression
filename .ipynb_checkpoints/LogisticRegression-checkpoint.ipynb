```python
missing_per_row = df.isnull().sum(axis=1)
rows_with_3_or_more_missing = df[missing_per_row >= 3]
print(rows_with_3_or_more_missing)
```

```python
df = df.drop([1122, 1747])
df = df.reset_index(drop=True)
```

```python
df[df["TenYearCHD"].isnull()]
```

```python
df.info()
```

```python
df.head()
```

```python
from sklearn.impute import SimpleImputer
```

```python
median_cols = ["education", "cigsPerDay", "totChol", "BMI", "heartRate", "glucose"]
median_imputer = SimpleImputer(strategy="median")
df[median_cols] = median_imputer.fit_transform(df[median_cols])
```

```python
df.info()
```

```python
df["BPMeds"].value_counts()
```

```python
mode_cols = ["BPMeds"]
mode_imputer = SimpleImputer(strategy="most_frequent")
df[mode_cols] = mode_imputer.fit_transform(df[mode_cols])
```

```python
print(df.isnull().sum())
```

```python
X=df.drop('TenYearCHD',axis=1)
y=df['TenYearCHD']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=15)
```

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
logistic.fit(X_train,y_train)
y_pred=logistic.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
score=accuracy_score(y_pred,y_test)
print("score: " , score)
print(classification_report(y_pred,y_test))
print("confusion matrix: \n" , confusion_matrix(y_pred,y_test))
```

```python
model=LogisticRegression()
penalty=['l1', 'l2', 'elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver=['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver)
```

```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
```

```python
from sklearn.model_selection import GridSearchCV

grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
grid
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
```

```python
score=accuracy_score(y_pred,y_test)
print("score: " , score)
print(classification_report(y_pred,y_test))
print("confusion matrix: \n" , confusion_matrix(y_pred,y_test))
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
model=LogisticRegression()
randomcv=RandomizedSearchCV(estimator=model,param_distributions=params,cv=5,scoring='accuracy')
```

```python
randomcv.fit(X_train,y_train)
```

```python
randomcv.best_score_
```

```python
randomcv.best_params_
```

```python
y_pred=randomcv.predict(X_test)
```

```python
score=accuracy_score(y_pred,y_test)
print("score: " , score)
print(classification_report(y_pred,y_test))
print("confusion matrix: \n" , confusion_matrix(y_pred,y_test))
```
